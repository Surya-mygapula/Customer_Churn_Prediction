```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("Churn_Modelling.csv")
dataset.head()
```

```python
X = dataset.iloc[:,3:-1]
y = dataset.iloc[:,-1].values
```

```python
X.isnull().sum()
X.dtypes
X=pd.get_dummies(X,drop_first=True)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.15 , stratify=y , random_state=0)
```

```python
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
import pickle
pickle.dump(rf, open('Customer_Churn_Prediction.pkl', 'wb'))
```

```python
new_data = pd.DataFrame({
    'Geography_France': [1],
    'Geography_Germany': [0],
    'Gender_Male': [0],
    'NumOfProducts': [2],
    'HasCrCard': [1],
    'IsActiveMember': [1]
})
```

```python
expected_columns = X.columns
for col in expected_columns:
    if col not in new_data.columns:
        new_data[col] = 0
new_data = new_data[expected_columns]
```

```python
new_data_scaled = sc.transform(new_data)
```

```python
new_prediction = rf.predict(new_data_scaled)
print("Predicted Churn:", "Churn" if new_prediction[0] == 1 else "No Churn")
```
